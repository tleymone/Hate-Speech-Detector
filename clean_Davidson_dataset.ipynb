```python
import pandas as pd
import re
```

```python
# Load the dataset using pandas
df = pd.read_csv('./davidson_labeled_data.csv', index_col=0)
```

```python
# 24783 lignes
df.shape
```

```python
df.tail()
```

```python
df.drop(columns=["count", "hate_speech", "offensive_language", "neither"], inplace=True)
```

```python
df.reset_index(drop=True, inplace = True)
```

```python
df.tail()
```

```python
df.sample(100)
```

```python
re.sub(r'&#[0-9]+', '', "This is a sample string with emojis like &#8220 and &#128512")
```

```python
for i in range(len(df.tweet)):
  t = df.tweet[i]

  # remove "RT"
  t = re.sub(r'\bRT\b',"", t)

  # remove mention @
  t = re.sub(r"@\S+", "", t)

  # remove URL
  t = re.sub("http[s]?\://\S+", "", t)

  # remove emoji "&#"
  t = re.sub(r'&#[0-9]+', '', t)

  # remove extra space
  t = re.sub('\s+',' ', t)

  df.tweet[i] = t
```

```python
df.to_csv("clean_davidson_labeled_data.csv")
```
